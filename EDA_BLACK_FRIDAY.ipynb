```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('train.csv')
```

3. **Gender and Purchase**: We can see from the dataset that it includes both male and female customers. 5. **Age and Occupation**: The dataset provides information about the age and occupation of the customers. Further analysis could be done to explore how different age groups and occupations impact the purchase behavior.

6. **Product Categories**: The dataset contains information about three product categories (Product_Category_1, Product_Category_2, and Product_Category_3). There are variations in product preferences, and further exploration can reveal insights into the most popular product categories among customers.

In conclusion, this exploratory data analysis gives us a preliminary understanding of the Black Friday dataset. Further in-depth analysis, feature engineering, and modeling could be performed to predict customer behavior, identify key factors influencing purchase decisions, and optimize marketing strategies to maximize sales on Black Friday.

# UNIVARIATE ANALYSIS

## Histogram of the 'Purchase' column

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Purchase'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.title('Purchase Amount Distribution')
plt.show()
```

We  can concude that the maximun purchase amount ranges between 7000 to 8000.

## Bar plot for 'Age' column

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', data=df, palette='viridis')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

We can conclude that the maximum age of customers ranges from 26 to 35 and the minimum from 0 to 17.

# BIVARIATE ANALYSIS

## Box plot of 'Purchase' against 'City_Category'

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='City_Category', y='Purchase', data=df, palette='magma')
plt.xlabel('City Category')
plt.ylabel('Purchase Amount')
plt.title('Purchase Amount by City Category')
plt.show()
```

##Bar  plot of 'Purchase' against 'Occupation'

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Occupation', y='Purchase', data=df, ci=None, palette='viridis')
plt.xlabel('Occupation')
plt.ylabel('Purchase Amount')
plt.title('Purchase Amount by Occupation')
plt.show()
```

In this plot we can see that the occupations are assigned certain numbers and according to that numbering system, the maximum purchase is from occupation 12.

#Categorical Variables Analysis

##  Pie chart for gender against purchase

```python
purchase_by_gender = df.groupby('Gender')['Purchase'].sum()

# Plot the pie chart
plt.figure()
colors = ['pink', 'skyblue']
plt.pie(purchase_by_gender, labels=purchase_by_gender.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Purchase Amount by Gender')
plt.show()
```

Here we can see  the majority of the customers are male.
